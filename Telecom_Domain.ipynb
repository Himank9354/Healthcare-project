```python
import os
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/Telcom Data.csv')
dataset.head()
```

```python
# Checking for imbalance data
```

```python
dataset['Churn'].value_counts()
```

```python
dataset['Churn'] = dataset['Churn'].replace({'Yes':1, "No":0})
```

```python
dataset.columns
```

```python
for i in dataset.columns:
  print("***********************************************",i,
        "*******************************************************")
  print()
  print(set(dataset[i].tolist()))
  print()
```

```python
# Using automation tool (pycaret)

from pycaret.classification import *
```

```python
# Spliting data into train test part

train_data = dataset.sample(frac = 0.7, random_state=101).reset_index(drop=True)
test_data = dataset.drop(train_data.index).reset_index(drop=True)
```

```python
print(train_data.shape, test_data.shape)
```

```python
exp_clf = setup(data=train_data, target='Churn', session_id=120)
```

```python
# Comparing all model
compare_models()
```

```python
# create a model
knn = create_model('knn')
```

```python
# HyperParameter Tunning

tuned_knn = tune_model(knn)
```

```python
# Evaluate the model
evaluate_model(tuned_knn)
```

```python
predict_model((tuned_knn))
```

```python
# Using test data

unseen_prediction = predict_model(knn, data=test_data)
unseen_prediction
```

```python
# save model
save_model(knn, 'knn_model')
```
